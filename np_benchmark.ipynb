## Benchmarking `numpy` Functions
- let's benchmark some fundamental `np` matrix and vector functions to get an understanding of the implementation speed
- we will connect this to the theoretical time complexity of these operations

```python
import numpy as np
from plot import Plot
P = Plot()
dims = 1
P.benchmark_np(np.linalg.norm,
               type_='vec',
               gen_info={'range':(2,25),'dims':dims,'rand':'base'},
               theory_info={
               'f':lambda x:(x**2)/(18*10**9), #18 * 10^9
               'prediction':2,
               'label':f'O(n^{2})'}
               )
```
